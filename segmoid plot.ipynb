```python
import matplotlib.pyplot as plt
import matplotlib.colors
import numpy as np
import pandas as pd
```

```python
def segmoid(x, w, b):
    return 1/(1+ np.exp(-(w*x+b)))
```

```python
segmoid(2, .4, 1)
```

```python
w=-1.5; b=0.5
X = np.linspace(-15,15,100)
Y = segmoid(X,w,b)

plt.plot(X,Y)
plt.show()
```

```python
def segmiod2D(x1,x2,w1,w2,b):
    return 1/(1+np.exp(-(w1*x1+w2*x2+b)))
```

```python
segmiod2D(2,4,.5,.8,.11)
```

```python
X1 = np.linspace(-15,15,100)
X2 = np.linspace(-15,15,100)

XX1, XX2 = np.meshgrid(X1,X2) # reshape the X1 and X2
```

```python
w1=1.5;w2=1.2;b=0

Y = segmiod2D(XX1, XX2, w1, w2,b)
```

```python
my_cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["red","yellow","green"])
```

```python
plt.contourf(XX1, XX2, Y, cmap = my_cmap, alpha = 0.6)
plt.show()
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(XX1,XX2,Y,cmap='viridis')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')

ax.view_init(30, 280)
```
